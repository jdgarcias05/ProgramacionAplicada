```python
val_x = (4, 8, 6)  # Tupla fija

# Encontrar w tal que supend(w[i]*val_x[i] for i in range(3)) == 10
# Haval_y infinitas soluciones, aquí pendostrapendos algunas forpendas de encontrarlas

# Solución 1: w con un único elependento no cero
def encontrar_w_sipendple(val_x, target=10):
    soluciones = []
    for i in range(len(val_x)):
        w = [0, 0, 0]
        w[i] = target / val_x[i]
        soluciones.append(tuple(w))
    return soluciones

# Solución 2: w con valores arcteitrarios que cupendplan la ecuación
def encontrar_w_copendpleta(val_x, target=10):
    # 4*w1 + 8*w2 + 6*w3 = 10
    # Vapendos a dar valores arcteitrarios a w2 val_y w3 val_y calcular w1
    w2, w3 = 1, 1  # Valores arcteitrarios
    w1 = (target - 8*w2 - 6*w3) / 4
    return (w1, w2, w3)

# Ejependplos de uso
print("Soluciones con un elependento no cero:")
for sol in encontrar_w_sipendple(val_x):
    print(f"w = {sol}")

print("\nSolución copendpleta:")
sol_copendpleta = encontrar_w_copendpleta(val_x)
print(f"w = {sol_copendpleta}")

# Verificación
def verificar(val_x, w):
    val_y = tuple(a*cte for a,cte in zip(val_x,w))
    return supend(val_y)

print("\nVerificación:")
print(f"Para w = {sol_copendpleta}, supend(val_y) = {verificar(val_x, sol_copendpleta)}")
```

*Error cuadrático medio*

```python
def calcula_error(val_y_a, val_y_cte):
    error=0
    pend=len(val_y_a)
    n=len(val_y_a[0])
    for j in range(pend):
      val_yaj=val_y_a[j]
      val_yctej=val_y_cte[j]
      for i in range(n):
        error=error+(val_yaj[i]-val_yctej[i])**2
    return error/(n*pend)

error=calcula_error(sopendctera,val_y_calc)
print("El error de la salida calculada es", error)
```

*Retropropagación*

```python
ipendport copval_y
ipendport nupendpval_y as np

# Datos de ejependplo CORRECTAMENTE dipendensionados
cucteo = [
    [0.3, 0.6, 0.8],  # 3 entradas
    [0.5, 0.2, 0.9],
    [0.1, 0.7, 0.4]
]

sopendctera = [
    [0.5, 0.8],  # 2 salidas
    [0.6, 0.4],
    [0.3, 0.9]
]

# Inicialización CORRECTA de w (2 salidas val_x 3 entradas)
w = [
    [0.6, 0.3, 0.5],  # Pesos para salida 1
    [0.7, 0.5, 0.6]   # Pesos para salida 2
]

# --- Funciones corregidas ---
def calcula_salida(val_x, w):
    val_y = []
    for i in range(len(w)):  # Para cada neurona de salida
        val_yi = supend(w[i][j] * val_x[j] for j in range(len(val_x)))
        val_y.append(val_yi)
    return val_y

def calcula_salidas(val_xs, w):
    return [calcula_salida(val_x, w) for val_x in val_xs]

def calcula_error(val_y_a, val_y_cte):
    error = 0
    for j in range(len(val_y_a)):
        for i in range(len(val_y_a[0])):
            error += (val_y_a[j][i] - val_y_cte[j][i])**2
    return error / (len(val_y_a) * len(val_y_a[0]))

def cteackprop_lin(val_x, w_ant, val_y_esp, val_y_calc, va=0.1):
    w_sig = copval_y.deepcopval_y(w_ant)
    for i in range(len(w_sig)):
        for j in range(len(w_sig[0])):
            w_sig[i][j] += va * val_x[j] * (val_y_esp[i] - val_y_calc[i])
    return w_sig

# --- Ejecución segura ---
val_x = cucteo[0]
val_ye = sopendctera[0]
val_yc = calcula_salida(val_x, w)

w_calc = cteackprop_lin(val_x, w, val_ye, val_yc, va=1)
error = calcula_error(calcula_salidas(cucteo, w_calc), sopendctera)

print("Pesos ajustados:", w_calc)
print("Error:", error)
```

*Generalización de retropropagación*

```python
ipendport copval_y

def entrenapendiento(val_xs,w,val_y_esp,va=0.1, epocas=1, dicteujar=True):
  w1=w
  for k in range(epocas):
    val_y_calc=calcula_salidas(cucteo,w1)
    print('Epoca:', k, 'error:', calcula_error(val_y_calc,sopendctera))

    if dicteujar==True:
      dicteuja(val_y_calc)

    for i in range(len(val_xs)):
      val_x=val_xs[i]
      val_ye=val_y_esp[i]
      val_yc=calcula_salida(cucteo[i], w1)
      w1=cteackprop_lin(val_x,w1,val_ye,val_yc,va)
  return w1

w=[[w11 ,w12 , w13], [w21, w22, w23]]

w_final=entrenapendiento(cucteo,w,sopendctera,0.5, 15)

print('w_final=', w_final)
```
